```python
import pandas as pd

# Cargar el archivo Excel
file_path = (r"C:\Users\usuario\Desktop\Ironhack Labs\Semana 22 (Proyecto Final) (06.08)\DataSetHotelesSanFrancisco.xlsx")  # Cambia esto a la ruta donde está tu archivo
data = pd.read_excel(file_path)

# Mostrar las primeras filas del DataFrame
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de los gráficos
sns.set(style="whitegrid")

# Distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(data['price2'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Precios de los Hoteles')
plt.xlabel('Precio (USD)')
plt.ylabel('Frecuencia')
plt.show()
```

1. Distribución de Precios de los Hoteles

1.a Descripción: Este gráfico es un histograma que muestra cómo se distribuyen los precios de los hoteles en el dataset.

2.a Interpretación:

- Las barras representan diferentes rangos de precios y su altura indica cuántos hoteles tienen precios dentro de ese rango.

- Esto te ayuda a identificar si la mayoría de los hoteles están en un rango de precios bajo, medio o alto.
Además, la curva de densidad (si usas kde=True) te da una idea de la distribución continua de los precios.

```python
# Relación entre precio y calificación
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price2', y='rating', data=data, hue='rating_title', palette='viridis', s=100)
plt.title('Relación entre Precio y Calificación')
plt.xlabel('Precio (USD)')
plt.ylabel('Calificación')
plt.legend(title='Rating Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

2. Relación entre Precio y Calificación

2.a Descripción: Este es un scatter plot (gráfico de dispersión) que compara el precio de cada hotel con su calificación.

2.b Interpretación:

- Cada punto en el gráfico representa un hotel.

- El eje X muestra los precios y el eje Y las calificaciones.

- Puedes observar si existe una correlación entre el precio y la calificación: por ejemplo, si los hoteles más caros tienden a tener mejores calificaciones.

- Los colores de los puntos indican el "título de calificación" (e.g., "Very good", "Superb"), lo que añade una capa adicional de interpretación.

```python
# Calificaciones promedio por ubicación
plt.figure(figsize=(14, 6))
location_mean_rating = data.groupby('location')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=location_mean_rating.values, y=location_mean_rating.index, palette='coolwarm')
plt.title('Calificación Promedio por Ubicación')
plt.xlabel('Calificación Promedio')
plt.ylabel('Ubicación')
plt.show()
```

3. Calificaciones Promedio por Ubicación

3.a Descripción: Este gráfico de barras horizontales muestra la calificación promedio de los hoteles agrupados por su ubicación en San Francisco.

3.b Interpretación:

- Cada barra representa una ubicación específica.

- La longitud de la barra indica la calificación promedio de los hoteles en esa área.

- Este gráfico es útil para comparar qué zonas de San Francisco tienen hoteles mejor calificados, lo cual podría ser importante si estás buscando analizar áreas específicas.

```python
# Limpiar la columna 'number_of_ratings'
data['number_of_ratings'] = data['number_of_ratings'].str.replace(' reviews', '').str.replace(',', '').astype(int)

# Relación entre número de reseñas y calificación
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_ratings', y='rating', data=data, hue='rating_title', palette='plasma', s=100)
plt.title('Relación entre Número de Reseñas y Calificación')
plt.xlabel('Número de Reseñas')
plt.ylabel('Calificación')
plt.legend(title='Rating Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

4. Relación entre Número de Reseñas y Calificación

4.a Descripción: Otro scatter plot que compara el número de reseñas que tiene un hotel con su calificación.

4.b Interpretación:

- El eje X muestra el número de reseñas, y el eje Y muestra la calificación del hotel.

- Este gráfico permite analizar si los hoteles con más reseñas tienden a tener calificaciones más altas o bajas.

- Los puntos coloridos representan diferentes títulos de calificación, ayudándote a ver cómo se distribuyen las opiniones en función del número de reseñas.
