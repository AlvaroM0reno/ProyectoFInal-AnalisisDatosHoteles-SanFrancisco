```python
import pandas as pd

# Cargar el archivo Excel
file_path = (r"C:\Users\usuario\Desktop\Ironhack Labs\Semana 22 (Proyecto Final) (06.08)\DataSetHotelesSanFrancisco.xlsx")  # Cambia esto a la ruta donde est√° tu archivo
data = pd.read_excel(file_path)

# Mostrar las primeras filas del DataFrame
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo de los gr√°ficos
sns.set(style="whitegrid")

# Distribuci√≥n de precios
plt.figure(figsize=(10, 6))
sns.histplot(data['price2'], bins=20, kde=True, color='skyblue')
plt.title('Distribuci√≥n de Precios de los Hoteles')
plt.xlabel('Precio (USD)')
plt.ylabel('Frecuencia')
plt.show()
```

1. Distribuci√≥n de Precios de los Hoteles

1.a Descripci√≥n: Este gr√°fico es un histograma que muestra c√≥mo se distribuyen los precios de los hoteles en el dataset.

2.a Interpretaci√≥n:

- Las barras representan diferentes rangos de precios y su altura indica cu√°ntos hoteles tienen precios dentro de ese rango.

- Esto te ayuda a identificar si la mayor√≠a de los hoteles est√°n en un rango de precios bajo, medio o alto.
Adem√°s, la curva de densidad (si usas kde=True) te da una idea de la distribuci√≥n continua de los precios.

```python
# Relaci√≥n entre precio y calificaci√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price2', y='rating', data=data, hue='rating_title', palette='viridis', s=100)
plt.title('Relaci√≥n entre Precio y Calificaci√≥n')
plt.xlabel('Precio (USD)')
plt.ylabel('Calificaci√≥n')
plt.legend(title='Rating Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

2. Relaci√≥n entre Precio y Calificaci√≥n

2.a Descripci√≥n: Este es un scatter plot (gr√°fico de dispersi√≥n) que compara el precio de cada hotel con su calificaci√≥n.

2.b Interpretaci√≥n:

- Cada punto en el gr√°fico representa un hotel.

- El eje X muestra los precios y el eje Y las calificaciones.

- Puedes observar si existe una correlaci√≥n entre el precio y la calificaci√≥n: por ejemplo, si los hoteles m√°s caros tienden a tener mejores calificaciones.

- Los colores de los puntos indican el "t√≠tulo de calificaci√≥n" (e.g., "Very good", "Superb"), lo que a√±ade una capa adicional de interpretaci√≥n.

```python
# Calificaciones promedio por ubicaci√≥n
plt.figure(figsize=(14, 6))
location_mean_rating = data.groupby('location')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=location_mean_rating.values, y=location_mean_rating.index, palette='coolwarm')
plt.title('Calificaci√≥n Promedio por Ubicaci√≥n')
plt.xlabel('Calificaci√≥n Promedio')
plt.ylabel('Ubicaci√≥n')
plt.show()
```

3. Calificaciones Promedio por Ubicaci√≥n

3.a Descripci√≥n: Este gr√°fico de barras horizontales muestra la calificaci√≥n promedio de los hoteles agrupados por su ubicaci√≥n en San Francisco.

3.b Interpretaci√≥n:

- Cada barra representa una ubicaci√≥n espec√≠fica.

- La longitud de la barra indica la calificaci√≥n promedio de los hoteles en esa √°rea.

- Este gr√°fico es √∫til para comparar qu√© zonas de San Francisco tienen hoteles mejor calificados, lo cual podr√≠a ser importante si est√°s buscando analizar √°reas espec√≠ficas.

```python
# Limpiar la columna 'number_of_ratings'
data['number_of_ratings'] = data['number_of_ratings'].str.replace(' reviews', '').str.replace(',', '').astype(int)

# Relaci√≥n entre n√∫mero de rese√±as y calificaci√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_ratings', y='rating', data=data, hue='rating_title', palette='plasma', s=100)
plt.title('Relaci√≥n entre N√∫mero de Rese√±as y Calificaci√≥n')
plt.xlabel('N√∫mero de Rese√±as')
plt.ylabel('Calificaci√≥n')
plt.legend(title='Rating Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

4. Relaci√≥n entre N√∫mero de Rese√±as y Calificaci√≥n

4.a Descripci√≥n: Otro scatter plot que compara el n√∫mero de rese√±as que tiene un hotel con su calificaci√≥n.

4.b Interpretaci√≥n:

- El eje X muestra el n√∫mero de rese√±as, y el eje Y muestra la calificaci√≥n del hotel.

- Este gr√°fico permite analizar si los hoteles con m√°s rese√±as tienden a tener calificaciones m√°s altas o bajas.

- Los puntos coloridos representan diferentes t√≠tulos de calificaci√≥n, ayud√°ndote a ver c√≥mo se distribuyen las opiniones en funci√≥n del n√∫mero de rese√±as.
